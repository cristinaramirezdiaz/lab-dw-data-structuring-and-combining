```python
import pandas as pd
```

```python
# Your code goes here
url1 = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv'
url2 = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv'
url3 = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv'
df = pd.read_csv(url1)
df2 = pd.read_csv(url2)
df3 = pd.read_csv(url3)

df
```

```python
df2
```

```python
df3
```

```python
df.columns = df.columns.str.lower().str.replace(" ", "_")
df2.columns = df2.columns.str.lower().str.replace(" ", "_")
df3.columns = df3.columns.str.lower().str.replace(" ", "_")
```

```python
df = df.rename(columns={
    "st": "state"
})
df
```

```python
df2 = df2.rename(columns={
    "st": "state"
})
df2
```

```python
df_total = pd.concat([df, df2, df3], axis=0)

df_total
```

```python
df_total.gender.value_counts()
```

```python
df_total['gender'].replace({'Male': 'M', 'female': 'M', 'Femal': 'M'}, inplace=True)

df_total['state'].replace({'WA': 'Washington', 'Cali': 'California', 'AZ': 'Arizona'}, inplace=True)

df_total.education.replace({'Bachelors': 'Bachelor'}, inplace=True)

df_total['vehicle_class'].replace({'Sports Car': 'Luxury', 'Luxury Car': 'Luxury', 'Luxury SUV': 'Luxury'}, inplace=True)

df_total['customer_lifetime_value'].replace({'%': ''}, regex=True, inplace=True)
```

```python
df_total.sample(10)
```

```python
df_total.customer_lifetime_value = df_total.customer_lifetime_value.astype(float)
df_total.number_of_open_complaints = df_total.number_of_open_complaints.str.split('/').str[1]
df_total.number_of_open_complaints = df_total.number_of_open_complaints.astype(float) # No puedo pasarlo a int porque hay NaN
```

```python
df_total.sample(10)
```

```python
df.isnull().sum() # isnull() sirve para ver si hay valores faltantes
```

```python
# Identificar las columnas con null o vacÃ­as
null_cols = df_total.columns[df_total.isnull().any()]
print(f"Columnas vacÃ­as o con valores nulos: {null_cols}")

# Separar las columnas categÃ³ricas y numÃ©ricas
cat_vars = df_total.select_dtypes(include=['object']).columns
num_vars = df_total.select_dtypes(include=['float64' , 'int64']).columns

# Cambiar los valores nulos con la moda en las variables categÃ³ricas

for x in null_cols:
    if x in cat_vars:
        df_total[x].fillna(df_total[x].mode()[0] , inplace = True)

# Lo mismo pero con la media en las numÃ©ricas

for x in null_cols:
    if x in num_vars:
        df_total[x].fillna(df_total[x].mean() , inplace = True)
```

```python
df_total.sample(10)
```

```python
df_total.duplicated().sum()
```

```python
df_total.drop_duplicates(inplace=True)

# Reiniciar el Ã­ndice

df_total.reset_index(drop = True, inplace= True)
```

1. You work at the marketing department and you want to know which sales channel brought the most sales in terms of total revenue. Using pivot, create a summary table showing the total revenue for each sales channel (branch, call center, web, and mail).
Round the total revenue to 2 decimal points.  Analyze the resulting table to draw insights.

```python
pivot_df = df.pivot_table(index='sales_channel', values=['total_claim_amount'], aggfunc='sum') # el aggfunc sirve para sumar los valores de dos paÃ­ses duplicados
pivot_df
```

```python
sales_total = pivot_df.round(2)
sales_total
```

2. Create a pivot table that shows the average customer lifetime value per gender and education level. Analyze the resulting table to draw insights.

```python
average_lifetime = df.pivot_table(index='gender', columns = 'education', values=['customer_lifetime_value'], aggfunc='mean') # el aggfunc sirve para sumar los valores de dos paÃ­ses duplicados
average_lifetime = average_lifetime.round(2)
average_lifetime
```
